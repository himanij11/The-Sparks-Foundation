```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style('whitegrid')
import plotly.express as px
```

```python
df = pd.read_csv('SampleSuperstore.csv')
df.head()
```

```python
df.dtypes
```

```python
df.isnull().sum()
```

```python
df.nunique()
```

```python
df.corr()
```

#### Profit and Sales have a positive correlation

```python
df.drop(['Postal Code'], axis = 1, inplace = True)
df
```

```python
df.duplicated().sum()
```

```python
df[df.duplicated()]
```

```python
df.drop_duplicates(inplace = True)
```

```python
df.duplicated().sum()
```

```python
df.sort_values('Profit', ascending = True, inplace = True)
df.head()
```

```python
sns.countplot(x='Segment',data = df)
```

#### Analysis - Here we can infer consumer has the highest count.

```python
sns.countplot(x='Ship Mode',data = df)
```

#### Analysis - Here we can infer Standard Class has the highest count and Same Day has the lowest count.

```python
sns.countplot(x='Segment', hue = 'Ship Mode',data = df)
```

#### Analysis - Here we can see Consumer with Standard Class as ship mode has the highest count while Corporate Segment with Same Day as ship mode has the lowest count.

```python
sns.heatmap(df.corr(), cmap = 'Blues_r', annot = True)
```

#### Analysis - Profit and Sales have a positive correlation while sales and discount have a negative correlation.

```python
sns.countplot(x = 'Region', data = df)
plt.show()
```

#### Analysis - West Region has the maximum count.

```python
sns.barplot(x = 'Category', y = 'Sales', data = df)
plt.show()
```

#### Analysis - Technology has the highest count.

```python
plt.figure(figsize=(18,10))
sns.countplot(x = 'State', data = df)
plt.xticks(rotation = 70)
plt.show()
```

#### Analysis - California has the highest count.

```python
sns.pairplot(df)
```

```python
plt.figure(figsize=(15, 5))
sns.barplot(x = 'Category', y = 'Sales', data = df, palette = 'flare')
plt.xticks(rotation = 20)
plt.show()
```

#### Analysis - Technology has the highest sales.

```python
plt.figure(figsize=(15,5))
sns.barplot(x = 'Sub-Category', y = 'Sales', data = df, palette = 'mako')
plt.xticks(rotation = 20)
plt.show()
```

#### Analysis - Copiers has the highest sales.

```python
plt.figure(figsize=(15, 5))
sns.barplot(x = 'Sub-Category', y = 'Profit', data = df, palette = 'GnBu_r')
plt.xticks(rotation = 20)
plt.show()
```

#### Analysis - Copiers has the highest profit.

```python
plt.figure(figsize=(15, 5))
sns.lineplot(x = 'Discount', y = 'Sales', data = df)
plt.show()
```

#### Analysis - As the discount increases the sales also increase. 

```python
sns.catplot(x = 'Category',y = 'Sub-Category', data = df)
plt.show()
```

#### Analysis - Here we can see the count of Category v/s Sub-Category.

```python
sns.scatterplot(x = 'Profit', y = 'Sales', data = df)
plt.show()
```

#### Analysis - A scatterplot of Profit v/s Sales.
